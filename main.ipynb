### Testing Loop

```python

import torch
from PokemonRedEnv import create_envs
from policy import Policy

RUN_MODE = "train" #train or eval
ROM_PATH = "PokemonRed.gb"
NUM_EVAL_EPISODES = 1
NUM_ENVS = 5
```

```python

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

envs = create_envs(num_envs=NUM_ENVS, rom_path=ROM_PATH, state_dir="StartingFiles",wind="null")# <---------------------"SDL2" or "null"
print("Initializing Dreamer Policy...")
policy = Policy(envs=envs, device=device, visualize_dreams=True)

if RUN_MODE == "train":

    
    
    print("Starting Training Mode...")
    policy.train()


elif RUN_MODE == "eval":
    print("Starting Evaluation Mode...")
    policy.evaluate(num_episodes=NUM_EVAL_EPISODES)
else:
    print(f"Invalid RUN_MODE: '{RUN_MODE}'. Please set it to 'train' or 'eval'.") 

envs.close()
print("All environments safely closed.")
```
