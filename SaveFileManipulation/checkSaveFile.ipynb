```python
import os
from pyboy import PyBoy

# Configuration
ROM_PATH = 'PokemonRed.gb'
STATE_PATH = 'PokemonRed.HighCity.state' # PokemonRed.Advanced.state

pyboy = PyBoy(ROM_PATH, cgb=True, sound_volume=0)


if os.path.exists(STATE_PATH):
    print(f"Loading existing state from {STATE_PATH}...")
    with open(STATE_PATH, "rb") as f:
        pyboy.load_state(f)
else:
    print(f"No save state found at {STATE_PATH}. Starting a new game session.")

print("Playing in COLOR... Close the window to save and exit.")

try:
    # pyboy.tick() returns True as long as the window is open
    while pyboy.tick():
        pass
finally:
    print(f"Closing... Saving game state to {STATE_PATH}")
    with open(STATE_PATH, "wb") as f:
        pyboy.save_state(f)
    
    # Properly shut down the emulator
    pyboy.stop()
```
